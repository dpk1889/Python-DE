```python
def palindrome4(s):
    n=len(s)
    first=0
    last=n-1
    while(first<last):
        if s[first]==s[last]:
            first+=1
            last-=1
        else:
         return False
        return True 
s='madam'
palindrome4(s)
```

```python
#characters counting 
def compress(s):
    n=len(s)
    new=''
    count=1
    for i in range(n-1):
        if s[i]==s[i+1]:
         count+=1
        else:
            new=new+s[i]+ str(count)
            count=1
    new=new+s[n-1]+str(count)
    return new
s='aaabbbaaaaadhhhgg'
print(compress(s))
    
    
    



```

```python
#fizzbuzz problem divisibility problem 
def fizzbuzz(n):
    for i in range(1,n+1):
        if(i%3==0 and i%5==0):
            print("divisible by 15 fizzbuzz")
        elif i%3==0:
            print("divisible by 3 fizz")
        elif i%5==0:
            print("divisible by 5 buzz")
        else : print(i)
n=25
fizzbuzz(n)

```

```python
def fizzbuzz(n):
    d={3:'fizz',5:'buzz'}
    for i in range(1,n+1):
        result=''
        for k,v in d.items():
            if i%k==0:
                result+=v
        if not result:
            result=i
        print(result)       
n=20
fizzbuzz(n)
```

```python
#character occurrence 
#least or max repeating 
# count particular element occurrence 
#count all character occurrence

def occur(s):
    ch={}
    for i in s:
        if i in ch:
         ch[i]=ch[i]+1
        else:
          ch[i]=1
    print(ch)
    resultmin=(min(ch,key=ch.get))
    resultmax=(max(ch,key=ch.get))

    print (resultmin)
    print(resultmax)
s='ASADCDVVJRNJRFFFR'   
occur(s)
 
```

```python
# upper lower conversion in a sentence 
def convert(s):
    l=[]
    temp=s.split("_")
    for i in temp:
        l.append(i[0].lower()+i[1:].upper())
    s=" ".join(l)
    print(s)
s="this_is_a_notebook"
convert(s)
```

```python
#armstrong number 
def arm(n):
    sum=0
    temp=n
    while temp>0:
        digit=temp%10
        sum=sum+(digit**3)
        temp=temp//10
    if n==sum:
     print(n)
n=371
arm(n)
```

```python
#armstrong number 
def arm(start,end):
    for n in range(start,end):
     sum=0
     temp=n
     while temp>0:
        digit=temp%10
        sum=sum+(digit**3)
        temp=temp//10
     if n==sum:
      print(n)
arm(0,400)
```

```python
# factorial of a number 
def fact(n):
    fact=1
    for i in range(1,n+1):
     fact=fact*i
    return fact   

fact(4)
    
```

```python
def factsum(n):
    temp=n
    sum=0
    while temp>0:
        digit=temp%10
        sum=sum+fact(digit)
        temp=temp//10
    if sum==n:
        print('yes')
    else:
        print('no')
factsum(145)        
        
```
